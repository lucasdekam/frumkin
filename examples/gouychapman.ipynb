```python
from frumkin.electrolyte import LatticeElectrolyte, Ion, Water
from frumkin.gongadze_iglic import GongadzeIglic
import numpy as np

potential = np.linspace(-0.4, 0.4, 200)
concentrations = [1e-3, 5e-3, 9e-3, 13e-3, 17e-3]
results = []

def get_model(conc):
    el = LatticeElectrolyte([
        Water(),
        Ion(r"Na$^+$", 15, conc, +1),
        Ion(r"Cl$^-$", 2, conc, -1),
    ])

    return GongadzeIglic(el, ohp=4) #, eps_stern=10)

for conc in concentrations:
    model = get_model(conc)
    result = model.voltammetry(potential, tol=1e-5)
    results.append(result)
```

```python
import matplotlib.pyplot as plt
%matplotlib widget
plt.style.use("../lucas.mplstyle")

fig, ax = plt.subplots(figsize=(4, 3))

for conc, result in zip(concentrations, results):
    ax.plot(result.potential, result.capacitance, label=f"{conc*1e3:.1f}mM")

# ax.set_ylim([0, 100])
ax.set_xlim([-0.4, 0.4])
ax.set_xlabel("Potential (V)")
ax.set_ylabel(r"$C$ ($\mu$F/cm$^2$)")
ax.legend(frameon=False)
fig.tight_layout()
fig.savefig("gi-cap.png")
```

```python
profiles = []
potentials = [-0.25, -0.5, -1]

for p in potentials:
    model = get_model(10e-3)
    prf = model.single_point(p, tol=1e-5)
    profiles.append(prf)
```

```python
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

colors = ['tab:red', 'tab:purple', 'tab:cyan']

for i, p in enumerate(potentials):
    prf = profiles[i]
    
    ax.plot(prf.x, prf.potential, label=r"$\phi_0=$" + f"{p:.1f} V", color=colors[i])
    ax2.plot(prf.x, prf.concentrations['Na$^+$'], color=colors[i])

# ax.set_ylim([0, 100])
# ax.set_xlim([-1, 1])
# ax.set_xlabel("Potential (V)")
# ax.set_ylabel(r"$C$ ($\mu$F/cm$^2$)")
ax.set_xlim([0, 20])
ax.set_xlabel(r"z ($\mathrm{\AA}$)")
ax.set_ylabel(r"$\phi$ (V)")
ax2.set_xlim([0, 20])
ax2.set_ylim([0, 6])
ax2.set_xlabel(r"z ($\mathrm{\AA}$)")
ax2.set_ylabel(r"$c_+$ (M)")
ax.legend(frameon=False)
fig.tight_layout()
fig.savefig("gi-prf.png")
```
