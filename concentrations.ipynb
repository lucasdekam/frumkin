```python
from frumkin.electrolyte import LatticeElectrolyte, Ion, Water
from frumkin.gongadze_iglic import GongadzeIglic
import numpy as np

potential = np.linspace(-1, 1, 200)
concentrations = [1e-3, 5e-3, 10e-3, 25e-3] 
# conc = 10e-3
# params = [-0.2, -0.1, 0, 0.1, 0.2]
results = []

for e in concentrations:
    el = LatticeElectrolyte([
        Water(min_eps=78),
        Ion(r"Na$^+$", 6, e, +1),
        Ion(r"Cl$^-$", 6, e, -1),
    ])
    model = GongadzeIglic(el, xmax=1000, delta_chemi=0.5)

    result = model.voltammetry(potential, tol=1e-6)
    results.append(result)
    
```

```python
import matplotlib.pyplot as plt
from scipy import constants 
plt.style.use('examples/frumkin.mplstyle')
%matplotlib widget

def frumkin_isotherm(e, onset, charge=1):
    n_ad = 0.139 
    temperature = 300
    beta = constants.elementary_charge / (constants.Boltzmann * temperature )
    theta_max = 0.1
    prefactor = n_ad * constants.elementary_charge / constants.angstrom ** 2 * charge * 100
    charge = prefactor * theta_max * np.exp(-beta*(onset - e)) / (1 + np.exp(-beta*(onset - e)))
    return np.gradient(charge, e)

fig = plt.figure(figsize=(4,3))
# ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot() #122, sharex=ax1)

for result, p in zip(results, concentrations):
    y = result.capacitance
    ax2.plot(potential, y, label=f"{p}")

ax2.set_xlim(left=potential[0], right=potential[-1])
ax2.set_ylim([0, 75])
ax2.legend(frameon=False)

fig.savefig('ariba-normal.png', dpi=200)
```
