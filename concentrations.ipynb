```python
from frumkin.electrolyte import LatticeElectrolyte, Ion, Water
from frumkin.gongadze_iglic import GongadzeIglic
import numpy as np

potential = np.linspace(-1, 1, 200)
concentrations = [1e-3, 5e-3, 10e-3, 15e-3] 
results = []

for conc in concentrations:
    el = LatticeElectrolyte([
        Water(min_eps=78),
        Ion(r"Na$^+$", 8, conc, +1),
        Ion(r"Cl$^-$", 8, conc, -1),
    ])
    model = GongadzeIglic(el, xmax=1000, delta_chemi=0.5)

    result = model.voltammetry(potential, tol=1e-6)
    results.append(result)
    
```

```python
import matplotlib.pyplot as plt
plt.style.use('examples/frumkin.mplstyle')
%matplotlib widget

fig = plt.figure(figsize=(5,3))
# ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot() #122, sharex=ax1)

for result, conc in zip(results, concentrations):
    result.plot(ax2, y='capacitance', label=f"{conc*1e3:.0f}mM")

ax2.set_xlim(left=potential[0], right=potential[-1])
ax2.set_ylim([0, 100])
ax2.legend(frameon=False)

fig.savefig('ariba-normal.png', dpi=200)
```

```python
from frumkin.tools.langevin import langevin_x
n_sites = 0.15
water_coverage = 0.5
dip = 1
kappa = model.kappa
eps_2 = 10

n_sites * water_coverage * dip * kappa * langevin_x(dip * 1) / eps_2
```
